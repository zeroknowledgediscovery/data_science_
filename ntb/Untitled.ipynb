```python
import pandas as pd
%matplotlib inline
import pylab as plt
TH=.1
D=25
```

```python
df=pd.read_csv('../modelprop.csv',names=['src','tgt','distance','gamma','delay','phase'])
```

```python
df=df[df.gamma<TH]

df['srcval']=[x.split('#')[4] for x in df.src]

df['tgtval']=[x.split('#')[4] for x in df.tgt]
```

```python
df0=df[(df.tgtval=='prec') & (df.distance<1005)].groupby(['delay']).mean()
```

```python
#df0.phase.plot()
ax=df0.gamma.ewm(alpha=.8).mean().plot(style='-k.',logx=False)
ax.set_xlim(0,10)
plt.figure()

ax=df0.phase.ewm(alpha=.9).mean().plot(style='-k.',logy=False,logx=False)
ax.set_xlim(0,10)
```

```python
df1=df[(df.distance<D) & (df.tgtval=='prec')].groupby(['distance']).mean()

#df0.phase.plot()
df1.gamma.ewm(alpha=.1).mean().plot(style='r.',logy=True,logx=True)
plt.figure()

df1.phase.ewm(alpha=.1).mean().plot(style='r.',logy=True,logx=True)

```

```python
df1=df[(df.distance<D) & (df.tgtval=='winter') & (df.srcval=='winter')].groupby(['distance']).mean()

#df0.phase.plot()
df1.gamma.ewm(alpha=.1).mean().plot(style='o',logy=False,logx=False)
plt.figure()

df1.phase.ewm(alpha=.1).mean().plot(style='o',logy=False,logx=False)



#df.groupby('tgt').max()
```

```python
df1=df[(df.distance<D) & (df.tgtval=='prec') & (df.srcval=='prec')].groupby(['distance']).mean()

#df0.phase.plot()
df1.gamma.ewm(alpha=.1).mean().plot(style='-k.',logy=False,logx=False)
plt.figure()

df1.phase.ewm(alpha=.1).mean().plot(style='-k.',logy=False,logx=False)

```
